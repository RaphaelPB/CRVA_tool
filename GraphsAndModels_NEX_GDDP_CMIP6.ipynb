```python
import scipy
dist_name = 'exp'

# ParamÃ¨tres de la loi
dist = getattr(scipy.stats, dist_name)

# ModÃ©liser la loi
param = dist.fit(precipitation_2050_projet0_ssp126_ACCESS_CM2)
```

```python
param
```

```python
dist
```

```python
loc = param[-2]
scale = param[-1]
arg = param[:-2]

pdf = dist.pdf(x, loc=loc, scale=scale, *arg)

plt.figure(figsize=(12,8))
plt.plot(x, pdf, label=dist_name, linewidth=3) 
plt.plot(x, y, alpha=0.6)
plt.legend()
plt.show()
```
